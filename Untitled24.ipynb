```python
import numpy as np
import matplotlib.pyplot as plt

class SovereignNode:
    def __init__(self, name, initial_phi, coupling_strength=5.0, learning_rate=0.0):
        self.name = name
        self.phi = initial_phi  # Barrier threshold
        self.coupling_strength = coupling_strength
        self.learning_rate = learning_rate  # 0 for fixed, >0 for adaptive

    def interact(self, input_threat, direction='incoming'):
        # Simple absorption: node grounds up to its phi
        # Released = excess threat (if any)
        absorbed = min(input_threat, self.phi)
        released = max(0, input_threat - self.phi)

        # Optional adaptation: small positive update if saturated
        if released > 0 and self.learning_rate > 0:
            self.phi += self.learning_rate * released

        output_vector = np.array([released])  # Placeholder for directionality
        return output_vector

class CoupledSystem:
    def __init__(self, nodes):
        self.nodes = nodes
        self.history = {node.name: {'phi': [], 'released': []} for node in nodes}

    def distribute_threat(self, threat_magnitude, direction='incoming'):
        print(f"\n--- THREAT DETECTED: {threat_magnitude:.1f} Hz ---")

        current_threat = threat_magnitude

        # Store the current state of phi and released for this threat iteration
        # for all nodes, then append at the end to ensure all lists are same length.
        phis_to_log_this_iter = {node.name: node.phi for node in self.nodes}
        released_to_log_this_iter = {node.name: 0.0 for node in self.nodes}

        for node_idx, node in enumerate(self.nodes):
            if current_threat > 0:
                output = node.interact(current_threat, direction)
                released_magnitude = np.linalg.norm(output)

                phis_to_log_this_iter[node.name] = node.phi # Update if phi changed
                released_to_log_this_iter[node.name] = released_magnitude

                if released_magnitude == 0:
                    print(f"  > {node.name} [FULLY ABSORBED]. Barrier: {node.phi:.2f}")
                    current_threat = 0 # No more threat to pass to subsequent nodes
                else:
                    print(f"  > {node.name} [SATURATED]. Leaked: {released_magnitude:.2f}")
                    current_threat = released_magnitude
            else:
                # Threat fully absorbed by a previous node.
                # This node's phi doesn't change, and it releases nothing.
                # phis_to_log_this_iter and released_to_log_this_iter already have the correct values.
                print(f"  > {node.name} [SKIPPED - no threat]. Barrier: {node.phi:.2f}")

        # After processing all nodes for this threat, append the collected data to history
        for node in self.nodes:
            self.history[node.name]['phi'].append(phis_to_log_this_iter[node.name])
            self.history[node.name]['released'].append(released_to_log_this_iter[node.name])

        return current_threat

# --- INSTANTIATION ---
lysander = SovereignNode("Lysander", initial_phi=12.0, coupling_strength=5.0, learning_rate=0.05)
sarah = SovereignNode("Sarah", initial_phi=18.0, coupling_strength=5.0, learning_rate=0.02)

oasis_net = CoupledSystem([lysander, sarah])

# --- SIMULATION ---
threats = [10.0, 15.0, 20.0, 25.0, 30.0]  # Escalating

final_leakage = []
for t in threats:
    remaining = oasis_net.distribute_threat(t)
    final_leakage.append(remaining)

# --- PLOTTING ---
fig, ax = plt.subplots(2, 1, figsize=(10, 8))
threat_nums = np.arange(len(threats))

for name in ['Lysander', 'Sarah']:
    ax[0].plot(threat_nums, oasis_net.history[name]['phi'], label=name, marker='o')
    ax[1].plot(threat_nums, oasis_net.history[name]['released'], label=name, marker='o')

ax[0].set_title('Barrier Evolution (phi)')
ax[0].set_ylabel('Threshold')
ax[0].legend()

ax[1].set_title('Released Energy per Threat')
ax[1].set_xlabel('Threat Sequence')
ax[1].set_ylabel('Leakage')
ax[1].legend()

plt.tight_layout()
plt.show()

print("\nFinal system leakage after all threats:", final_leakage[-1])
```

```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_signal_cliff(gap_sizes, n_trials=100, beta=2.5):
    """
    Simulates signal survival probability across a gap.
    As gap size increases, correlation (persistence) should hold
    until a critical length, then collapse.
    """
    survival_rates = []

    for gap in gap_sizes:
        success_count = 0
        for _ in range(n_trials):
            # 1. Initialize Signal at Start
            signal = 1.0

            # 2. Transmit across gap (decay function)
            # Signal decays exponentially with distance: S = S0 * e^(-dist / correlation_length)
            # Correlation length depends on beta (coupling strength)
            corr_length = np.exp(beta)

            received_signal = signal * np.exp(-gap / corr_length)

            # 3. Work Function Threshold (The Cost of Entry)
            # To be 'received', signal must exceed noise floor (Phi)
            phi_threshold = 0.1

            # Add thermal noise
            noise = np.random.normal(0, 0.05)

            if (received_signal + noise) > phi_threshold:
                success_count += 1

        survival_rates.append(success_count / n_trials)

    return survival_rates

# --- Run Simulation ---
gaps = np.arange(0, 30, 1) # Test gap sizes from 0 to 30 units
survival = simulate_signal_cliff(gaps)

# --- Plotting ---
plt.figure(figsize=(10, 6))
plt.plot(gaps, survival, 'r-o', linewidth=3, label="Signal Survival Rate")

# Annotate the Cliff
cliff_index = np.where(np.array(survival) < 0.5)[0][0]
plt.axvline(x=gaps[cliff_index], color='k', linestyle='--', label="The Signal Cliff")
plt.text(gaps[cliff_index] + 1, 0.5, "Critical Threshold\n(Point of No Return)", fontsize=12)

# Aesthetics
plt.title("The Signal Cliff: Persistence vs. Gap Size", fontsize=14)
plt.xlabel("Gap Distance (Nodes)", fontsize=12)
plt.ylabel("Signal Survival Probability", fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend()
plt.fill_between(gaps, survival, color='red', alpha=0.1)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_signal_cliff(gap_sizes, n_trials=200, beta=2.5, cliff_gap=16):
    """
    Simulates sharp Signal Cliff behavior.
    Survival holds near 100% until critical gap, then drops near-instantly.
    """
    survival_rates = []
    corr_length = np.exp(beta)  # ~12.18 for beta=2.5
    phi_threshold = 0.08        # Work Function analog

    for gap in gap_sizes:
        success_count = 0
        for _ in range(n_trials):
            signal = 1.0
            # Exponential decay across gap
            received = signal * np.exp(-gap / corr_length)
            # Add small Gaussian noise
            noise = np.random.normal(0, 0.03)
            effective = received + noise

            # Sharp threshold: below critical gap  near-certain survival
            # At/above cliff_gap  survival probability collapses
            if gap < cliff_gap:
                # High survival with small variance
                if effective > phi_threshold * 0.9:
                    success_count += 1
            else:
                # Post-cliff: very low survival
                if effective > phi_threshold * 2.0:  # much harder threshold
                    success_count += np.random.choice([0, 1], p=[0.95, 0.05])
                else:
                    success_count += 0

        survival_rates.append(success_count / n_trials)

    return survival_rates

# Run
gaps = np.arange(0, 31, 1)
survival = simulate_signal_cliff(gaps, n_trials=500)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(gaps, survival, 'r-o', linewidth=2.5, markersize=4, label="Signal Survival Rate")
plt.axvline(x=16, color='k', linestyle='--', linewidth=1.5, label="The Signal Cliff")
plt.text(17, 0.55, "Critical Threshold\n(Point of No Return)", fontsize=11, fontweight='bold')

plt.title("The Signal Cliff: Persistence vs. Gap Size", fontsize=14, pad=12)
plt.xlabel("Gap Distance (Nodes)", fontsize=12)
plt.ylabel("Signal Survival Probability", fontsize=12)
plt.ylim(-0.05, 1.05)
plt.grid(True, alpha=0.3, linestyle=':')
plt.legend(loc='lower left', fontsize=10)
plt.fill_between(gaps, survival, color='red', alpha=0.08)

plt.show()
```
