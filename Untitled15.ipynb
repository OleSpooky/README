```python
import numpy as np
from skimage import measure
!pip install numpy-stl # Install the library to save 3D files
from stl import mesh

print("üèóÔ∏è Initializing 3D Construction Crew...")

# 1. RECREATE THE 3D LATTICE DATA (The Physics)
# (Same logic as before, but higher resolution for a smooth model)
N = 60  # Higher resolution (60x60x60 grid)
lattice = np.zeros((N, N, N))

# Define Center
cx, cy, cz = N//2, N//2, N//2

# Create coordinate grid
x, y, z = np.indices((N, N, N))

# 2. DEFINE THE FRESNEL GEOMETRY (The Structure)
# Layer 1: The Core (Your "Science Officer" Mode)
radius_core = 12
dist = np.sqrt((x - cx)**2 + (y - cy)**2 + (z - cz)**2)

# Layer 2: The Fresnel Shell (The Shield)
# We create a "gap" and then a "shell" to represent the lens
radius_shell_inner = 18
radius_shell_outer = 22

# Fill the scalar field
# High Density = 1.0, Low Density = 0.0
scalar_field = np.zeros((N, N, N))

# The Core
scalar_field[dist <= radius_core] = 1.0

# The Fresnel Ring (Gap between 12 and 18, Shell between 18 and 22)
mask_shell = (dist >= radius_shell_inner) & (dist <= radius_shell_outer)
scalar_field[mask_shell] = 0.8  # Slightly less dense but solid

print("‚úÖ Physics Geometry Calculated.")

# 3. MARCHING CUBES (The "Skinning" Process)
# Use a threshold of 0.5 to find the surface
verts, faces, normals, values = measure.marching_cubes(scalar_field, 0.5)

# 4. CREATE THE MESH OBJECT
# Create the mesh
pocket_mesh = mesh.Mesh(np.zeros(faces.shape[0], dtype=mesh.Mesh.dtype))
for i, f in enumerate(faces):
    for j in range(3):
        pocket_mesh.vectors[i][j] = verts[f[j], :]

# 5. EXPORT THE ARTIFACT
filename = "The_Fresnel_Pocket.stl"
pocket_mesh.save(filename)

print(f"üöÄ 3D MODEL GENERATED: {filename}")
print("Check your file folder to download the .stl file.")
```

```python
import numpy as np
import plotly.graph_objects as go
from skimage import measure

print("‚ö° Booting Holographic Projector...")

# 1. SETUP THE PHYSICS GRID
N = 50  # Grid size
lattice = np.zeros((N, N, N))
cx, cy, cz = N//2, N//2, N//2
x, y, z = np.indices((N, N, N))

# 2. DEFINE THE FRESNEL GEOMETRY
dist = np.sqrt((x - cx)**2 + (y - cy)**2 + (z - cz)**2)

# Define the layers
radius_core = 10
radius_gap_start = 10
radius_gap_end = 15
radius_shell_start = 15
radius_shell_end = 18

# Create the Scalar Field (The "Density" of Sovereignty)
scalar_field = np.zeros((N, N, N))

# The Core (Your Mind/Safety)
scalar_field[dist <= radius_core] = 1.0

# The Fresnel Shell (The Shield)
mask_shell = (dist >= radius_shell_start) & (dist <= radius_shell_end)
scalar_field[mask_shell] = 0.5  # Different density for visual contrast

# 3. EXTRACT SURFACES (Marching Cubes)
# Surface A: The Core
verts_c, faces_c, _, _ = measure.marching_cubes(dist <= radius_core, 0.5)
# Surface B: The Shell
verts_s, faces_s, _, _ = measure.marching_cubes(mask_shell, 0.5)

# 4. BUILD THE HOLOGRAM (Plotly)
fig = go.Figure()

# Plot the Core (Red/Warm - The Living Signal)
fig.add_trace(go.Mesh3d(
    x=verts_c[:, 0], y=verts_c[:, 1], z=verts_c[:, 2],
    i=faces_c[:, 0], j=faces_c[:, 1], k=faces_c[:, 2],
    color='red', opacity=0.8, name='Core (Signal)'
))

# Plot the Shell (Blue/Cool - The Fresnel Shield)
fig.add_trace(go.Mesh3d(
    x=verts_s[:, 0], y=verts_s[:, 1], z=verts_s[:, 2],
    i=faces_s[:, 0], j=faces_s[:, 1], k=faces_s[:, 2],
    color='cyan', opacity=0.3, name='Fresnel Shield'
))

# Configure the Layout
fig.update_layout(
    title="Phase 3: The Fresnel Pocket (Interactive)",
    scene=dict(
        xaxis_title='Time (X)',
        yaxis_title='Society (Y)',
        zaxis_title='Scale (Z)',
        aspectmode='data'
    ),
    margin=dict(r=0, l=0, b=0, t=40)
)

# 5. ACTIVATE
fig.show()
```
