```python
import numpy as np
import matplotlib.pyplot as plt

# --- THE FRACTAL LATTICE ---
class CoupledSystem:
    def __init__(self, nodes):
        self.nodes = nodes
        self.history = {node.name: {'phi': [], 'released': []} for node in nodes}

    def distribute_threat(self, threat_magnitude, direction):
        """
        The threat hits the Network, not just the Node.
        Energy flows to the path of least resistance (Highest Barrier).
        """
        print(f"\n--- THREAT DETECTED: {threat_magnitude}Hz ---")

        # In a coupled system, the "Effective Barrier" is the sum of the local potentials
        # plus the network coupling. We simplify this as a "Shared Shield".
        # If Lysander is overwhelmed, Sarah absorbs the excess.

        total_released = 0
        current_threat = threat_magnitude

        for node in self.nodes:
            # Each node attempts to ground the signal
            # The 'Input' to the node is the remaining threat
            output = node.interact(current_threat, direction)
            released_magnitude = np.linalg.norm(output)

            # Log Data
            self.history[node.name]['phi'].append(node.phi)
            self.history[node.name]['released'].append(released_magnitude)

            if released_magnitude == 0:
                print(f"  > {node.name} [ABSORBED]. Barrier: {node.phi:.2f} + {node.c}")
                current_threat = 0 # Threat neutralized
                break
            else:
                print(f"  > {node.name} [SATURATED]. Leaked: {released_magnitude:.2f}")
                current_threat = released_magnitude # Pass the leakage to the next node

        return current_threat

# --- INSTANTIATING THE HETEROGENEITY ---

# 1. Lysander (Kinetic / Reactive)
# High Learning Rate (Alpha), Lower Initial Threshold
lysander = SovereignNode("Lysander", initial_phi=12.0, coupling_strength=5.0, learning_rate=0.15)

# 2. Sarah (Structural / Persistent)
# Low Learning Rate (Stable), High Initial Threshold (The Mandate)
sarah = SovereignNode("Sarah", initial_phi=18.0, coupling_strength=5.0, learning_rate=0.01)

# The Network
oasis_net = CoupledSystem([lysander, sarah])

# The Same "Sonic Breach" Scenario (Escalating to 20Hz)
threats = [10.0, 15.0, 20.0, 25.0]
vector = np.array([1, 0])

for t in threats:
    residual = oasis_net.distribute_threat(t, vector)
    if residual > 0:
        print(f"RESULT: SYSTEM FAILURE. Global Leakage: {residual:.2f}")
    else:
        print(f"RESULT: SYSTEM STABLE. Zero Release.")

# --- VISUALIZING THE HANDOFF ---
plt.figure(figsize=(10, 6))
plt.plot(oasis_net.history['Lysander']['phi'], label='Lysander (Kinetic Phi)', marker='o', color='blue')
plt.plot(oasis_net.history['Sarah']['phi'], label='Sarah (Structural Phi)', marker='s', color='green')
plt.title('Heterogeneous Stability: How the "We" Absorbs the "I"')
plt.xlabel('Threat Iterations')
plt.ylabel('Sovereign Potential (Hz)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np

class SovereignNode:
    def __init__(self, name, initial_phi, coupling_strength, learning_rate):
        self.name = name
        self.phi = initial_phi        # The Ethical Threshold (The Shield)
        self.c = coupling_strength    # The Lattice Connection (The We)
        self.alpha = learning_rate    # The History/Adaptation (The Memory)
        self.history = []

    def interact(self, observer_state_magnitude, direction_vector):
        """
        Executes the Unified Release Operator.

        Args:
            observer_state_magnitude (float): ||Os|| - The stimulus/threat intensity.
            direction_vector (np.array): d_hat - The vector of intent.

        Returns:
            np.array: E_released - The vector of resulting action.
        """
        # 1. Calculate the Effective Barrier (Phi + C)
        effective_threshold = self.phi + self.c

        # 2. Determine Surplus (||Os|| - Phi - C)
        surplus = observer_state_magnitude - effective_threshold

        # 3. Apply Rectified Linear Unit (ReLU) Logic
        # Max(0, surplus)
        magnitude_released = max(0, surplus)

        # 4. Calculate Vector Release
        e_released = direction_vector * magnitude_released

        # 5. Execute Threshold Update Rule (The Learning)
        # Phi(t+1) = Phi(t) + alpha * max(0, ||Os|| - Phi)
        # Note: The update is driven by the suprathreshold excess relative to Phi alone,
        # representing the internal stress on the agent.
        internal_stress = max(0, observer_state_magnitude - self.phi)
        self.phi += self.alpha * internal_stress

        # Log the state for the notebook
        self.history.append({
            "stimulus": observer_state_magnitude,
            "threshold": self.phi,
            "released": magnitude_released
        })

        return e_released

# --- Simulation Verification ---
# Let's run the "Sonic Breach" scenario through your equation.

# 1. Initialize Lysander (The Node)
# Phi=12.0 (The Mandate), C=5.0 (Sarah's Support), Alpha=0.1 (Learning)
lysander = SovereignNode("Lysander", initial_phi=12.0, coupling_strength=5.0, learning_rate=0.1)

# 2. The Attack (The Consortium)
# A series of attacks increasing in intensity: 10Hz, 15Hz, 20Hz
attacks = [10.0, 15.0, 20.0]
direction = np.array([1, 0]) # Directional attack

print(f"--- SIMULATION START: {lysander.name} ---")
for i, attack in enumerate(attacks):
    output = lysander.interact(attack, direction)
    print(f"Time {i}: Input {attack}Hz | Barrier {lysander.phi:.2f}+{lysander.c} | Released {np.linalg.norm(output):.2f} | New Phi {lysander.phi:.2f}")
```
